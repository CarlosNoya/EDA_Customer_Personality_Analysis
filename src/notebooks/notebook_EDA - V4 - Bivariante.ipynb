##### 8. EDUCATION - INCOME

```python
# EDUCATION - INCOME
pinta_distribucion_categoricas(df, ["Education"], relativa=False, mostrar_valores=True)
pinta_distribucion_categoricas(df, ["Education"], relativa=True, mostrar_valores=True)
plot_combined_graphs(df, ["Income"], whisker_width=2, bins = 40) 
plot_combined_graphs(no_outlier, ["Income"], whisker_width=2, bins = 40) 


plot_categorical_numerical_relationship(df, "Education", "Income", measure="mean", show_values=True)

plot_grouped_histograms(df, "Education", "Income", 3,)
plot_grouped_histograms(no_outlier, "Education", "Income", 3)
```

```python
# ANOVA
from scipy import stats

grupos = df['Education'].unique()  
income_by_education = [df[df['Education'] == grupo]['Income'] for grupo in grupos] 

f_val, p_val = stats.f_oneway(*income_by_education) 
print("Valor F:", f_val)
print("Valor p:", p_val)
```

Como P-Value <<<0.05 rechazamos hipotesis nula y por tanto reafirmamos la contrario (la hipotesis)

- Hay relación entre las variables EDUCATION e INCOME

#### Nota
- Histogramas de Graduate y Postgraduate son similares, sin embargo Undergraduate es distinto (se ha visto con y si el outlier máximo en ingresos)
- Los ingresos medios para "estudios superiores" son mayores que para aquellos que tienen estudios "basicos". Además, a mayor nivel educativo, mayores ingresos medios
